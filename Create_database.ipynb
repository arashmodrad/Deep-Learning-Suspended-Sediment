```python
def Copier(LL):
    for Loop in range(0, len(LL)):
        print('Item '+str(Loop+1)+':')
        folderPath = os.path.join(LL[Loop], 'Processed_images') 
        #print(folderPath)
        img_list = []
        for filename in listdir(folderPath):
            if filename.endswith('.npy'):
                try:
                    img_list.append(filename)
                    #img = Image.open('./'+filename) # open the image file
                    #img.verify() # verify that it is, in fact an image
                except (IOError, SyntaxError) as e:
                    print('Bad file:', filename) # print out the name
        #print( img_list)
        Experiment = folderPath.split('/')[2][:1]
        Experiment_type = folderPath.split('/')[3].split(' ')[0]
        Concentration = LL[Loop].split('/')[4]
        
        # gets the current working dir
        src_dir = os.getcwd() 
        # defining the dest directory
        dest_folder = os.path.join(src_dir, 'ModelDatabase2')
        
        if Experiment == '1':
            df = pd.read_excel('./LabData/1-North of Three Forks North-05.24.2020-05.27.2020/Records.xlsx', sheet_name=Experiment_type)     
            Con_Level = df[df.columns[4]]
            SSC_Level = df[df.columns[5]]
            TEMP_Level = df[df.columns[10]]
            
            found = Con_Level.str.contains(Concentration, case=False, na=False)
            found_ind = found.index[found == True].tolist()[0]
            
            Con_Value = Con_Level[found_ind]
            SSC_Value = SSC_Level[found_ind]
            TEMP_Value = TEMP_Level[found_ind]
            Meta = dict({'Con_Value': Con_Value, 'SSC_Value': SSC_Value, 'TEMP_Value': TEMP_Value})
            
            for image in img_list:
                Meta_name = image[:-4]+'_'+Experiment+'_'+Experiment_type+'_'+Concentration+'.json'
                Meta_path = os.path.join(dest_folder, Meta_name)
                
                with open(Meta_path, 'w', encoding='utf-8') as f:
                    json.dump(Meta, f, ensure_ascii=False, indent=4)
                    
                oldname = os.path.join(folderPath, image)
                Proper_Name = image[:-4]+'_'+Experiment+'_'+Experiment_type+'_'+Concentration+'.npy'
                newname = os.path.join(dest_folder, Proper_Name)
                shutil.copy(oldname, newname)
                
        elif Experiment == '2':
            df = pd.read_excel('./LabData/2-HRB-Clay-05.27.2020-05.28.2020/Records.xlsx', sheet_name=Experiment_type)     
            Con_Level = df[df.columns[4]]
            SSC_Level = df[df.columns[5]]
            TEMP_Level = df[df.columns[10]]
            
            found = Con_Level.str.contains(Concentration, case=False, na=False)
            found_ind = found.index[found == True].tolist()[0]
            
            Con_Value = Con_Level[found_ind]
            SSC_Value = SSC_Level[found_ind]
            TEMP_Value = TEMP_Level[found_ind]
            Meta = dict({'Con_Value': Con_Value, 'SSC_Value': SSC_Value, 'TEMP_Value': TEMP_Value})
                
            for image in img_list:
                Meta_name = image[:-4]+'_'+Experiment+'_'+Experiment_type+'_'+Concentration+'.json'
                Meta_path = os.path.join(dest_folder, Meta_name)
                with open(Meta_path, 'w', encoding='utf-8') as f:
                    json.dump(Meta, f, ensure_ascii=False, indent=4)
                    
                oldname = os.path.join(folderPath, image)
                Proper_Name = image[:-4]+'_'+Experiment+'_'+Experiment_type+'_'+Concentration+'.npy'
                newname = os.path.join(dest_folder, Proper_Name)
                shutil.copy(oldname, newname)
                
        elif Experiment == '4':
            df = pd.read_excel('./LabData/4 - Hat Ranch Soil - 6142021/Records.xlsx', sheet_name=Experiment_type)     
            Con_Level = df[df.columns[4]]
            SSC_Level = df[df.columns[5]]
            TEMP_Level = df[df.columns[10]]
            
            found = Con_Level.str.contains(Concentration, case=False, na=False)
            found_ind = found.index[found == True].tolist()[0]
            
            Con_Value = Con_Level[found_ind]
            SSC_Value = SSC_Level[found_ind]
            TEMP_Value = TEMP_Level[found_ind]
            Meta = dict({'Con_Value': Con_Value, 'SSC_Value': SSC_Value, 'TEMP_Value': TEMP_Value})
                
            for image in img_list:
                Meta_name = image[:-4]+'_'+Experiment+'_'+Experiment_type+'_'+Concentration+'.json'
                Meta_path = os.path.join(dest_folder, Meta_name)
                with open(Meta_path, 'w', encoding='utf-8') as f:
                    json.dump(Meta, f, ensure_ascii=False, indent=4)
                    
                oldname = os.path.join(folderPath, image)
                Proper_Name = image[:-4]+'_'+Experiment+'_'+Experiment_type+'_'+Concentration+'.npy'
                newname = os.path.join(dest_folder, Proper_Name)
                shutil.copy(oldname, newname)
        else:
            df = pd.read_excel('./LabData/3-BurnedSoil6.2.2021/Records.xlsx', sheet_name='Forward')     
            Concentration_2 = LL[Loop].split('/')[5]
            Experiment_type_2 = folderPath.split('/')[3].split(' ')[0]
            Concentration = Concentration_2
            Experiment_type = Experiment_type_2
            Con_Level = df[df.columns[4]]
            SSC_Level = df[df.columns[5]]
            TEMP_Level = df[df.columns[10]]
            
            found = Con_Level.str.contains(Concentration_2, case=False, na=False)
            found_ind = found.index[found == True].tolist()[0]
            
            Con_Value = Con_Level[found_ind]
            SSC_Value = SSC_Level[found_ind]
            TEMP_Value = TEMP_Level[found_ind]
            Meta = dict({'Con_Value': Con_Value, 'SSC_Value': SSC_Value, 'TEMP_Value': TEMP_Value})
                
            for image in img_list:
                Meta_name = image[:-4]+'_'+Experiment+'_'+Experiment_type_2+'_'+Concentration_2+'.json'
                Meta_path = os.path.join(dest_folder, Meta_name)
                with open(Meta_path, 'w', encoding='utf-8') as f:
                    json.dump(Meta, f, ensure_ascii=False, indent=4)
                    
                oldname = os.path.join(folderPath, image)
                Proper_Name = image[:-4]+'_'+Experiment+'_'+Experiment_type_2+'_'+Concentration_2+'.npy'
                newname = os.path.join(dest_folder, Proper_Name)
                shutil.copy(oldname, newname)
                
                
        print(Experiment+'_'+Experiment_type+'_'+Concentration+' DONE!')
        print('\n')
```
